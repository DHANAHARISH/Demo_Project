# Outlier remove 


```python
q1=df['Volume'].quantile(0.25)
q3=df['Volume'].quantile(0.75)
IQR=q3-q1
lower_range=q1-1.5*IQR
upper_range=q3+IQR 

print(f"q1 = {q1}\nq3 = {q3}\nIQR = {IQR}\nlower_range = {lower_range}\nupper_range = {upper_range}")
```

```python
df['Volume']=np.where(df['Volume']<lower_range,lower_range,
                      np.where(df['Volume']>upper_range,upper_range,
                      df['Volume']))
df['Volume']
```

```python
plt.figure(figsize=(8 ,5))
sns.boxplot(df['Volume'],orient='h')
plt.title('outlier romoved volume plot ')
```

```python
df['Year'] =df['Date'].dt.year
df['Month'] =df['Date'].dt.month
df['Day'] = df['Date'].dt.day
X=df.drop(['Date','Close','Adj Close'],axis=1).copy()
Y =df['Close'].copy()
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=345)
```

# Model plottree and elbow plot 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.tree import plot_tree
rf=RandomForestRegressor()
rf.fit(X_train,Y_train)
```

```python
# estimators_ is DecisionTreeRegressor make tree plot in random forest 
rf.estimators_
```

```python
plt.figure(figsize=(20,20))
plot_tree(rf.estimators_[0],max_depth=5,filled=True,fontsize=(15),feature_names=X.columns)
plt.show()
```

```python
# feature_importances_
plt.barh(X.columns,rf.feature_importances_)
```

```python
y_pred=rf.predict(X_test)
print('MSE:',mean_squared_error(Y_test,y_pred))
print('Train score ',rf.score(X_train,Y_train)*100)
print('Test score ',rf.score(X_test,Y_test)*100)
```

```python
def rf_function(k):
    rf=RandomForestRegressor(n_estimators=k)
    rf.fit(X_train,Y_train)
    y_pred=rf.predict(X_test)
    mse=mean_squared_error(Y_test,y_pred)
    return mse
```

```python
rf_function(4)
```

```python
k_value=range(1,21)
mse=[]
for k in k_value:
    error=rf_function(k)
    mse.append(error)
```

```python
mse
```

```python
plt.plot(k_value,mse,marker='o')
plt.xlabel('k')
plt.ylabel('error')
plt.title('rf elbow method')
plt.show()
```
